# Removing Unwanted Labels
This is just a short notebook that removes any labels that are furthest away from the median structure size. We do this to ensure that the only labels in the image are the masks of the structures themselves.

```python
from skimage import io
import os
import numpy as np

# needs to be changed to repository directory
home_directory = r"C:\Users\savill\OneDrive\Documents\PhD Jesse\Embryonic_organoid_prediction\Processing dataset for Paper\TLS_2D_morphostate_investigation/"
folder = home_directory + r"image_data\Processed Data\Masks Brightfield\72h"

mask_collection = io.imread_collection(folder+"/*.tif")
file_names = [os.path.split(fname)[1] for fname in mask_collection.files]

io.imshow(mask_collection[162])
```

```python
import organoid_prediction_python as opp
edited_masks = opp.keep_labels_closest_to_stack_median(mask_collection)

io.imshow(edited_masks[162])
```

```python
for name, mask in zip(file_names,edited_masks):
    io.imsave(f"{folder}/{name}",mask.astype("uint8"))
```
