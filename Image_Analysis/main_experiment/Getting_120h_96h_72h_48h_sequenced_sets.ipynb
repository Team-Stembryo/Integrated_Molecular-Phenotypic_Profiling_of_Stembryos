# Adding new Morphological Data to Sequenced Samples
Here we are adding our extracted features to tables with identifiers for sequenced samples. This is to allow us to add the features to R-Objects for the integrated analysis. First we load the data then we process each timepoint:

```python
import numpy as np
import numba
import pandas as pd

home_directory = r"C:\Users\savill\OneDrive\Documents\PhD Jesse\Embryonic_organoid_prediction\Processing dataset for Paper\TLS_2D_morphostate_investigation/"

table_location = home_directory + "tabular_data"


old_data_48 = pd.read_csv(f"{table_location}/OLD_Multi_BC_Variation_With_Morpho_sequenced_samples_48h.csv")
old_data_72 = pd.read_csv(f"{table_location}/OLD_Multi_BC_Variation_With_Morpho_sequenced_samples_72h.csv")
old_data_96 = pd.read_csv(f"{table_location}/OLD_Multi_BC_Variation_With_Morpho_sequenced_samples_96h.csv")
old_data_120= pd.read_csv(f"{table_location}/OLD_Multi_BC_Variation_With_Morpho_sequenced_samples_120h.csv")
old_data_72.head()
```

```python
recalculated_features = pd.read_csv(f"{table_location}/All_Samples_Feature_Space_All_Features_48h,72h,96h.csv",).drop("Unnamed: 0",axis=1)
recalculated_features
```

```python
filtered_48 = recalculated_features[(recalculated_features["Run"]=="RR")&(recalculated_features["Plate"]==2)].drop(
    ["Run","Plate"],
    axis=1
).rename(columns={"ID":"Sample"}).set_index("Sample")

feats_to_keep = ["Sample","Multi_BC","Multi_BC_#","Plate","Morphotype"]

samples_48h_seq_w_morpho_data = pd.concat(
    [old_data_48[feats_to_keep].set_index("Sample"),filtered_48.loc[old_data_48["Sample"].to_list()]],
    axis=1
)

samples_48h_seq_w_morpho_data.head()
```

## 72h

```python
filtered_72 = recalculated_features[(recalculated_features["Run"]=="RR")&(recalculated_features["Plate"]==3)].drop(
    ["Run","Plate"],
    axis=1
).rename(columns={"ID":"Sample"}).set_index("Sample")

feats_to_keep = ["Sample","Multi_BC","Multi_BC_#","Plate","Morphotype"]

samples_72h_seq_w_morpho_data = pd.concat(
    [old_data_72[feats_to_keep].set_index("Sample"),filtered_72.loc[old_data_72["Sample"].to_list()]],
    axis=1
)

samples_72h_seq_w_morpho_data.head()
```

## 96h

```python
filtered_96 = recalculated_features[(recalculated_features["Run"]=="RR")&(recalculated_features["Plate"]==5)].drop(
    ["Run","Plate"],
    axis=1
).rename(columns={"ID":"Sample"}).set_index("Sample")

feats_to_keep = ["Sample","Multi_BC","Multi_BC_#","Plate","Morphotype_96h"]

samples_96h_seq_w_morpho_data = pd.concat(
    [old_data_96[feats_to_keep].set_index("Sample"),filtered_96.loc[old_data_96["Sample"].to_list()]],
    axis=1
)

samples_96h_seq_w_morpho_data.head()
```

## 120h

```python
filtered_120 = recalculated_features[(recalculated_features["Run"]=="RR")&(recalculated_features["Plate"]==1)].drop(
    ["Run","Plate"],
    axis=1
).rename(columns={"ID":"Sample"}).set_index("Sample")

feats_to_keep = ["Sample","Multi_BC","Multi_BC_#","Plate","Morphotype_96h"]

samples_120h_seq_w_morpho_data = pd.concat(
    [old_data_120[feats_to_keep].set_index("Sample"),filtered_120.loc[old_data_120["Sample"].to_list()]],
    axis=1
)

samples_120h_seq_w_morpho_data.head()
```

Now we save the data:

```python
for hour,df in zip(
    ["048","072","096","120"],
    [
        samples_48h_seq_w_morpho_data,
        samples_72h_seq_w_morpho_data,
        samples_96h_seq_w_morpho_data,
        samples_120h_seq_w_morpho_data,
    ]):
    df.to_csv(f"{table_location}/Morphodata_All_Features_Sequenced_{hour}h.csv")
```
