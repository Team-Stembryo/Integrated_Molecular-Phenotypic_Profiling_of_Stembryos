# Combining All Features
In this notebook we combine the data from the python feature extraction as well as the cellprofiler feature extraction together. We start with the morphological data:

```python
import numpy as np
import numba
import pandas as pd

home_directory = r"C:\Users\savill\OneDrive\Documents\PhD Jesse\Embryonic_organoid_prediction\Processing dataset for Paper\TLS_2D_morphostate_investigation/"

table_location = home_directory + "tabular_data"

simple_morphologies=[]
straight_morphologies=[]
for hour in [48,72,96]:    
    simple_morphologies.append(pd.read_csv(f"{table_location}/{hour}h/simple_shape_features_{hour}h.csv").set_index(["Run","Plate","ID"]))
    straight_morphologies.append(pd.read_csv(f"{table_location}/{hour}h/straight_morgana_shape_features_{hour}h.csv").set_index(["Run","Plate","ID"]))

simple_morphologies[0]
```

Renaming columns to match our feature naming convention:

```python
def rename_morphologies(dataframes_list,morph_prefix):
    hour_prefixes = ["048h","072h","096h"]

    out_df_list = []
    for hour,df in zip(hour_prefixes,dataframes_list):
        mapping = {key:f"{hour}_{morph_prefix}_BF_AreaShape_{key}" for key in df.keys()}
        out_df_list.append(df.rename(columns=mapping))
    
    return out_df_list

simple_renamed = rename_morphologies(simple_morphologies,morph_prefix="raw")
pd.concat(simple_renamed,axis=1)
```

```python
straight_renamed = rename_morphologies(straight_morphologies,morph_prefix="str")
pd.concat(straight_renamed,axis=1)
```

Renaming the columns to match our naming convention:

```python
def rename_columns(df,hour,raw_str_prefix,channel_prefix,cellprof_remove = "__72h_brachyury_original"):
    new_keys = [key.replace(cellprof_remove,"") for key in df.keys()]
    mapping = {key:f"{hour}_{raw_str_prefix}_{channel_prefix}_{new_key}" for key,new_key in zip(df.keys(),new_keys)}
    return df.rename(columns = mapping)

renamed_bra_mask_72 = rename_columns(data_bra_mask_72,"072h","raw","CH_bra_mask")
renamed_bf_mask_72 = rename_columns(data_bf_mask_72,"072h","raw","CH__bf_mask")
renamed_bf_mask_72
```

Converting BRA pixel measurements into um measurements:

```python
pixel_size = 2.7493
length_columns = ["Axis","Perimeter","Diameter","Radius"]
bra_mask_columns = [column for column in df.columns if "CH_bra_mask_AreaShape" in column] 
for column in bra_mask_columns:
    if any([length in column for length in length_columns]):
        #print(column)
        df[column] = df[column].to_numpy() * pixel_size
    if ("Area" in column.replace("_AreaShape_","")) and ("fraction" not in column):
        print(column)
        df[column] = df[column].to_numpy() * (pixel_size**2)
df.head()
```
