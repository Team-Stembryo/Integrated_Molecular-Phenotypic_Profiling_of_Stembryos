```python
non_tl_data = pd.read_csv(f"{table_location}/All_Samples_Feature_Space_All_Features_48h,72h,96h.csv")

better_names = []
original_names = []
for key in non_tl_data.keys():
    if "CH" in key and "AreaShape" in key:
        mod_key = key[31:]
        if "Polarisation" in key or "Fraction" in key:
            mod_key = key[26:]
            
        mod_key = mod_key.replace("_"," ")
        mod_key = re.sub(r"(\w)([A-Z])", r"\1 \2", mod_key)
        mod_key = mod_key.title()

        mod_key = mod_key.replace("Locoefa","LoCo EFA")
        if key[5:].startswith("str"):
            mod_key = key[5:8] + " " + mod_key
        
        if "Polarisation" in key:
            mod_key = key[:4] + f"-[{mod_key}]-TmCH"
        else:
            mod_key = key[:4] + f"-[{mod_key}]-TmCH"
        
        print(key,mod_key)
        original_names.append(key)
        better_names.append(mod_key)
        
translation_frame_CH_shape_tp = pd.DataFrame({"original_names":original_names,"plot_names":better_names})
translation_frame_CH_shape_tp
```

```python
better_names = []
original_names = []
for key in non_tl_data.keys():
    if "CH" in key and "AreaShape" not in key and "bf_mask" in key and "Location" not in key:
        if "k_Intensity" in key:
            mod_key = key[31:]
        if "RadialDistribution" in key:
            mod_key = key[21:]
            mod_key = mod_key.replace("FracAtD","Sum")
            mod_key = mod_key.replace("MeanFrac","Mean")
        if "Polarisation" in key:
            mod_key = key[26:]
        mod_key = mod_key.replace("_"," ")
        mod_key = re.sub(r"(\w)([A-Z])", r"\1 \2", mod_key)
        mod_key = key[:4] + f"-[{mod_key}]-TmCH(BF)"
        print(mod_key)
        original_names.append(key)
        better_names.append(mod_key)

translation_frame_intensity_bf_mask_tp = pd.DataFrame({"original_names":original_names,"plot_names":better_names})
translation_frame_intensity_bf_mask_tp     
```
